```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Scores by School Spending

  * Overall Passing Rate (Average of the above two)

```python
# Reading Score Bins

spending_bins = [0, 600, 650, 700]
spending_bin_names = [">600", "$601-650", "$651-700"]

school_data_complete['Spending'] = pd.cut(school_data_complete['budget']/school_data_complete['size'], spending_bins, labels = spending_bin_names)

# Group data by size bins
scores_by_spending = school_data_complete.groupby("Spending")

# Make Avg/percentage values 


avg_math_sum = scores_by_spending['math_score'].mean()
pass_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('Spending')['Student ID'].count()/scores_by_spending['Student ID'].count()*100

avg_reading_sum = scores_by_spending['reading_score'].mean()
pass_read = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('Spending')['Student ID'].count()/scores_by_spending['Student ID'].count()*100

overall = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('Spending')['Student ID'].count()*100

# Create Data Frame

School_Spending_Summary_Table = pd.DataFrame({
   "Average Math Score": avg_math_sum,  
   "Average Reading Score": avg_reading_sum,
    "% Passing Math": pass_math,
   "% Passing Reading": pass_read,
   "Overall Passing": overall  })



School_Spending_Summary_Table.head()
```
